# Processing data

# step 1
+ return a DataFrame towns and the states from in the : university_towns.txt


+ consider data
        + in the rows : Name[edit][..] => this is States
        + in the rows : Name2(... university)[...] => this is Region Name

```python
data = []
sate = None
state_towns = []
with open('university_towns.txt') as file:
    # consider the each rows
    for line in file:
        thisline = line[:-1]
        
        # find State
        # the rows contain States , we can look string : '[edit]' in 6 the last characters
        if thisline[-6:] == '[edit]':
            # we take the characters from the first to the character in location : -6
            state = thisline[: -6]
            continue
        
        # find Region
        # under the each rows contain State ,we can look string : '(... university)' 
        # this the row contain region of State
        if '(' in line:
            town = thisline[: thisline.index('(') -1]
            state_towns.append([state, town])
        else:
            town = thisline
            state_towns.append([state, town])
        data.append(thisline)
    df_State_towns = pd.DataFrame(state_towns, columns = ['State', 'RegionName'])
    
    
df_State_towns     
```

# ask 4 
+ 'First creates new data showing the decline or growth of housing prices between the recession start and the recession bottom

+ then runs a ttestcomparing the university town values to the non-university towns values, return whether the alternative hypothesis (that the two groups are the same) is true or not as well as the p-value of the confidence. 

```python
# step 1 : create new data showing  between the recession start and recession bottom

# we have time of recession start (2008q3) and recession bottom(2009q2)
data = Convert_housing_data.copy()

# take the columns '2008q3'->'2009q2'
data = data.loc[:, '2008q3':'2009q2']
data = data.reset_index()
data.head(2)
```

```python
'''# the way 1
# compute prices ratio in during recssion
def prices_ration():
    ratio = (data['2008q3'] - data['2009q2'])/data['2008q3']
    return ratio
data['up&down'] = prices_ration()
data'''
```

```python
# the way 2 
def prices_ration(row):
    return ((row['2008q3'] - row['2009q2'])/row['2008q3'])

data['up&down'] = data.apply(prices_ration, axis =1)
data
```

```python
# we  take set RegionName from : df_State_towns
uni_town = df_State_towns['RegionName']
uni_town = set(uni_town)
uni_town
```

```python
# check in : data 
def is_uni_town(row):
    if row['RegionName'] in uni_town:
        return 1
    else:
        return 0
data['is_uni'] = data.apply(is_uni_town, axis = 1)
data.head(2)
```

```python
# we just consider during in the recssion 2008-2009
# classification the RegionName have university and RegionName have't university

# we take the values of the column 'up&down' , term have containing 'RegionName' (of df_State_towns) = 1
is_uni = data[data['is_uni'] == 1].loc[:, 'up&down'].dropna()

# we take the values of the column 'up&down' , term don't contain 'RegionName' (of df_State_towns) = 0
not_uni = data[data['is_uni'] == 0].loc[:, 'up&down'].dropna()
```

```python
from scipy.stats import ttest_ind
def better():
    if not_uni.mean() < is_uni.mean():
        return "non-university towns"
    else:
        return "university towns"
    
p_val = list(ttest_ind(not_uni, is_uni))[1]
p_val
```

```python
result = (True,p_val,better())
result
```
