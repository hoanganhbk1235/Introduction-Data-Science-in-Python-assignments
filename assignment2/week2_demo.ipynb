```python
import numpy as np
import pandas as pd
```

# Energy Indicators.xls :
+ which is a list of indicators (-chỉ số) of energy supply(-cung cấp) and renewable (-tái tạo) electricity production from the United Nations for the year 2013
+ keep in mind (-hãy nhớ rằng) that this an Excel file , and a not comma (-dấu phẩy) separated(-phân tách, phân chia ) values file
+ the first two columns are unneccessary (-không cần thiết) , so you should rid(-loại bỏ) get of them
+ and you should change the column labels so that the columns are: ['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable's]

```python
# file : xlsx read with : pd.read_excel
# file : xls read with : pd.ExcelFile
x = pd.ExcelFile('Energy Indicators.xls')
x
```

```python
energy = x.parse(skiprows = 17, skip_footer = (38))
energy.head(2)
```

```python
energy = energy[['Unnamed: 1','Petajoules','Gigajoules','%']]
energy.head(2)
```

```python
energy.columns = ['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']
energy
```

```python
# convert 'Energy Supply' and 'Energy Supply per Capita' to gigajoules (there are 1,000,000 gigajoules in a petajoule)

# For all countries which have missing (thiếu) data (e.g. data with "...") make sure this is reflected(-phản ánh) as np.NaN values.
```

```python
# For all countries which have missing (thiếu) data (e.g. data with "...") make sure this is reflected(-phản ánh) as np.NaN values.
energy[['Energy Supply', 'Energy Supply per Capita', '% Renewable']] =  energy[['Energy Supply', 'Energy Supply per Capita', 
                                            '% Renewable']].replace('...',np.NaN).apply(pd.to_numeric)
```

```python
# convert 'Energy Supply' and 'Energy Supply per Capita' to gigajoules (there are 1,000,000 gigajoules in a petajoule)
energy['Energy Supply'] = energy['Energy Supply']*1000000
```

```python
energy.head(2)
```

```python
# rename the following list of countries special
energy['Country'] = energy.Country.replace({"Republic of Korea": "South Korea",
"United States of America": "United States",
"United Kingdom of Great Britain and Northern Ireland": "United Kingdom",
"China, Hong Kong Special Administrative Region": "Hong Kong"})
```

```python
idx = np.where(energy.Country == 'United Kingdom')
ids = int(idx[0])
# access to the row through index 
ans = energy.iloc[ids, :].to_frame().T
#ans['Country']
ans
```

```python
#There are also several (-1 vài) countries with parenthesis(-dấu ngoặc đơn) in their name. Be sure to remove these
# remove the parenthesis and charactor special in the several name countries
# xóa cả cụm trong ngoặc đơn
# or the way 2
energy['Country'] = energy['Country'].str.replace(r"\(.*\)","")
```
