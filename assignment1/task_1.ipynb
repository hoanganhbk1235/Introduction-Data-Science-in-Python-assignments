```python
import pandas as pd
import numpy as np
```

we loads olympic dataset (olympic.csv) , which was derrived from the Wikipedia on All time Olympic Games Medals , and does some basic data cleaning 

```python
df_olympic = pd.read_csv('olympics.csv', index_col = 0, skiprows = 1)
df_olympic.head(2)
```

```python
df_olympic.tail(2)
```

```python
# then we consider (-xem xet) the each columns of data frame : df_olympic
# run tthe each column of df_olympic
for col in df_olympic.columns:
    # the each column convert to the corresponding (- tuong ung) name : Gold , Silver, Bronzer
    if col[:2] == '01':
        # example : col = 01! : have 3 character
        # we consider the 2 first character == 01 
        # we convert = Gold + col[4:](take from solution 4 of row ) => col = Gold 
        df_olympic.rename(columns = {col: 'Gold' + col[4:]} , inplace = True)
    if col[:2] == '02':
        df_olympic.rename(columns = {col: 'Silver' + col[4:]} , inplace = True)
    if col[:2] == '03':
        df_olympic.rename(columns = {col: 'Bronzer' + col[4:]} , inplace = True)
    if col[:1] == '№':
        df_olympic.rename(columns = {col :'#' + col[2:]} , inplace = True)
df_olympic.head(2)
    
```

```python
# split the index by '('
name_ids = df_olympic.index.str.split('\s\(')
df_olympic.index = name_ids.str[0]
df_olympic.head(2)
```
