# Question 5 :
+ Which the state has the most counties in it ?
+ this function should return single value

```python
# consider df_census : SUMLEV = 50 , corresponding CTYNAME is County (Counties)
df_counties = df_census[df_census.SUMLEV == 50]
df_counties.head(2)
```

```python
df = df_counties.groupby('STNAME')['CTYNAME'].count().to_frame()
df.head()
```

```python
a = max(df.CTYNAME)
a
```

```python
ans = df[df.CTYNAME == a]
ans.index.tolist()
```

```python
# the way 2
def most_counties():
    # ans : is type Series
    ans = df_counties.groupby('STNAME')['CTYNAME'].count()
    return ans.idxmax()
most_counties()
```

```python
# the way 2 
def three_most_pop_state():
    # take 3 counties the most population of the each state : dataframe
    y = df_counties.sort_values(by = ['STNAME','CENSUS2010POP'], ascending= False).groupby('STNAME').head(3)
    # compute sum population of 3 the most counties for pop of the each state : dataframe
    x = y.groupby('STNAME').sum().sort_values(by ='CENSUS2010POP', ascending = False)
    return x.head(3).index.tolist()
three_most_pop_state()
```
